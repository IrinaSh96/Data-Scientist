#База данных студентов#

```python
import sqlite3 as sq3
```

```python
con = sq3.connect('students.db')
cursor = con.cursor()
```

```python
cursor.execute('''CREATE TABLE students_list(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    surname TEXT NOT NULL,
    name TEXT NOT NULL,
    name_group_id INTEGER,
    FOREIGN KEY (name_group_id) REFERENCES groups (id)
)''')

cursor.execute('''CREATE TABLE groups(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name_group TEXT NOT NULL,
    name_faculty_id INTEGER,
    FOREIGN KEY (name_faculty_id) REFERENCES facultets (id)
)''')

cursor.execute('''CREATE TABLE facultets(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name_faculty TEXT NOT NULL,
    count_subjects INTEGER NOT NULL
)''')
con.commit()
```

```python
cursor.execute("SELECT * FROM sqlite_master")
cursor.fetchall()
```

##количество студентов каждого факультета##

```python
cursor.execute("""SELECT name_faculty as 'Название факульета', COUNT(surname)
    FROM students_list
    JOIN groups ON groups.id= students_list.name_group_id
    JOIN facultets ON facultets.id= groups.name_faculty_id
    GROUP BY name_faculty
""")
cursor.fetchall()
```
